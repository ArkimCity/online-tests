- 개요
    - 기존의 sorted 된 interval들이 들어있는 리스트에 newInterval 을 합쳐 리턴

- 시간복잡도
    - 첫 answer_intervals 생성시 intervals 의 길이 O(N)
    - not has_newInterval_inserted 의 경우 sorted 가 실행됩니다.
        - python sorted 는 Timsort 알고리즘을 사용한다고 함 - O(NlogN)

    - 종합 O(NlogN)

```python
class Solution:

    def is_overlapping(self, interval: List[int], other_interval: List[int]) -> bool:
        return (
            interval[0] <= other_interval[0] <= interval[1]
            or interval[0] <= other_interval[1] <= interval[1]
        ) or (
            other_interval[0] <= interval[0] <= other_interval[1]
            or other_interval[0] <= interval[1] <= other_interval[1]
        )

    def merge_if_overlaps(self, interval: List[int], newInterval: List[int]) -> List[int]:
        return [
            min(interval[0], newInterval[0]),
            max(interval[1], newInterval[1])
        ]

    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:

        has_newInterval_inserted = False

        answer_intervals = []
        for interval in intervals:

            # 오버랩 할 경우 기존 값을 확장해주고
            adjusted_inveral = interval
            if self.is_overlapping(interval, newInterval):
                adjusted_inveral = self.merge_if_overlaps(interval, newInterval)
                has_newInterval_inserted = True

            # 확장한 새로운 리스트들이 오버랩 할경우, 서로 합쳐준다.
            if len(answer_intervals) > 0 and self.is_overlapping(answer_intervals[-1], adjusted_inveral):
                answer_intervals[-1] = self.merge_if_overlaps(answer_intervals[-1], adjusted_inveral)
            else:
                answer_intervals.append(adjusted_inveral)

        # newInterval 이 한번도 들어가지 못했다면 위치에 맞게 들어가줘야 합니다.
        if not has_newInterval_inserted:
            answer_intervals = sorted(answer_intervals + [newInterval])

        return answer_intervals
```

2차

- 시간복잡도
    - 첫 answer_intervals 생성시 intervals 의 길이 O(N)
    - not has_newInterval_inserted 의 경우 answer_intervals 를 한번 iterate 하는 O(N)
        - 위치를 찾으면 insert 실행 -> for 문 안쪽에 있지만 단일 실행임으로 곱이 아니라 합으로 간주해도 괜찮음

    - 종합 O(N)

```python
class Solution:

    def is_overlapping(self, interval: List[int], other_interval: List[int]) -> bool:
        return (
            interval[0] <= other_interval[0] <= interval[1]
            or interval[0] <= other_interval[1] <= interval[1]
        ) or (
            other_interval[0] <= interval[0] <= other_interval[1]
            or other_interval[0] <= interval[1] <= other_interval[1]
        )

    def merge_if_overlaps(self, interval: List[int], newInterval: List[int]) -> List[int]:
        return [
            min(interval[0], newInterval[0]),
            max(interval[1], newInterval[1])
        ]

    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:

        has_newInterval_inserted = False

        answer_intervals = []
        for interval in intervals:

            # 오버랩 할 경우 기존 값을 확장해주고
            adjusted_inveral = interval
            if self.is_overlapping(interval, newInterval):
                adjusted_inveral = self.merge_if_overlaps(interval, newInterval)
                has_newInterval_inserted = True

            # 확장한 새로운 리스트들이 오버랩 할경우, 서로 합쳐준다.
            if len(answer_intervals) > 0 and self.is_overlapping(answer_intervals[-1], adjusted_inveral):
                answer_intervals[-1] = self.merge_if_overlaps(answer_intervals[-1], adjusted_inveral)
            else:
                answer_intervals.append(adjusted_inveral)

        # newInterval 이 한번도 들어가지 못했다면 위치에 맞게 들어가줘야 합니다.
        if not has_newInterval_inserted:
            if len(answer_intervals) == 0:
                return [newInterval]

            if newInterval[1] < answer_intervals[0][0]:
                return [newInterval] + answer_intervals

            for i, answer_interval in enumerate(answer_intervals):
                if i == len(answer_intervals) - 1:
                    return answer_intervals + [newInterval]

                next_answer_interval = answer_intervals[i + 1]
                if answer_interval[1] < newInterval[0] and newInterval[1] < next_answer_interval[0]:
                    answer_intervals.insert(i + 1, newInterval)
                    break

        return answer_intervals
```
